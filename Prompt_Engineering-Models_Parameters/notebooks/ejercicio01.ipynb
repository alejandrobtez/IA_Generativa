# Exploración de Técnicas de Prompt Engineering

## Configuración de variables y modelo

```python
%pip install openai python-dotenv
```

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

# Usamos la clase OpenAI estándar para endpoints de Proyecto de AI Foundry
client = OpenAI(
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),
    base_url=os.getenv("AZURE_OPENAI_ENDPOINT")
)

def ask_llm(prompt):
    try:
        # En AI Foundry, el 'model' es el deployment name
        deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT_NAME")
        
        response = client.chat.completions.create(
            model=deployment_name,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.7
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"❌ Error: {e}"

# Prueba de conexión
print("--- Probando conexión con Endpoint de Proyecto ---")
print(ask_llm("Hola, confirma si la conexión es exitosa."))
```

## 1.1 Selección de 6 Técnicas

### 1. Rol / Persona 🎭

**Definición:** técnica de asignar una identidad o perfil específico al modelo antes de darle una instrucción. Al definir quién debe ser el modelo, logramos que ajuste automáticamente su vocabulario, tono y nivel de detalle al contexto profesional deseado.

**Prompt:**

```python
prompt_rol = "Eres un experto en scouting de fútbol con 20 años de experiencia. Explica los riesgos de fichar a un chico de 10 años proveniente del futbol somalí para jugar en el Atlético de Madrid"
print(ask_llm(prompt_rol))
```

### 2. Few-shot 📚
**Definición:** Consiste en proporcionar ejemplos de entrada y salida dentro del prompt para que el modelo aprenda por imitación. Es la mejor forma de "entrenar" al modelo en segundos para que siga un formato, estilo o lógica específica que es difícil de explicar solo con instrucciones.

**Prompt:**

```python
prompt_few_shot = """Convierte lenguaje coloquial en lenguaje formal.

Ejemplo 1:
Entrada: 'Oye, pásame ese informe cuando puedas.'
Salida: 'Le agradecería que me enviara el informe en cuanto le sea posible.'

Ejemplo 2:
Entrada: 'No me mola nada ese diseño.'
Salida: 'Considero que el diseño actual no cumple con las expectativas requeridas.'

Entrada: 'Dime qué onda con el proyecto.'
Salida:"""
print(ask_llm(prompt_few_shot))
```

### 3. Chain-of-Thought 🔗
**Definición:** Consiste en pedirle al modelo que razone paso a paso antes de dar la respuesta final. Esta técnica mejora la precisión en tareas lógicas o matemáticas porque obliga al modelo a descomponer problemas complejos en partes pequeñas, evitando que cometa errores por "saltar" directamente a una conclusión precipitada.

**Prompt:**

```python
prompt_cot = "Si tengo 3 manzanas y compro 2 cajas que traen 6 manzanas cada una, pero 4 están podridas y las tiro, ¿cuántas manzanas tengo al final? Razona tu respuesta paso a paso."
print(ask_llm(prompt_cot))
```

### 4. Cues / Priming 🎯
**Definición:** Esta técnica consiste en escribir el inicio de la respuesta para "empujar" al modelo a seguir una estructura determinada. Al dejar el prompt abierto con una palabra, un signo o un formato parcial, condicionamos al modelo para que complete la información siguiendo exactamente ese patrón desde el primer token.

**Prompt:**

```python
prompt_cue = """Enumera los 3 mejores equipos de futbol que vistan de rojo y blanco.

Respuesta:
1. """
print(ask_llm(prompt_cue))
```

### 5. Format Forcing (JSON) 🔒
**Definición:** Es la técnica de obligar al modelo a entregar la respuesta en un formato específico, como JSON, XML o Markdown.

**Prompt:**

```python
prompt_json = "Analiza el sentimiento del texto: 'Estoy muy feliz con mi nuevo coche'. Devuelve el resultado en este formato JSON: { 'sentimiento': string, 'puntuacion': float (0-1) }"
print(ask_llm(prompt_json))
```

### 6. Give the model an 'out' 🚪
**Definición:** Consiste en darle al modelo una vía de escape explícita (como decir "No lo sé" o "Información no encontrada").

**Prompt:**

```python
prompt_out = "¿Quién ganó la liga de fútbol en el planeta Marte en 2024? Si no conoces la respuesta o la información no existe, responde: 'Datos no disponibles'."
print(ask_llm(prompt_out))
```

## 1.2 Casos de Uso Reales

### Caso 1: Extracción de entidades de texto legal
**Técnicas usadas:** Delimitadores + Format Forcing.
**Justificación:** El uso de delimitadores ayuda a separar el texto de las instrucciones, y el forzado de formato permite integrar la salida en una base de datos.

```python
prompt_caso1 = """Extrae los nombres de personas y empresas del siguiente texto.

Texto: ### El contrato firmado entre Juan Pérez y la empresa TechNova S.A. establece una duración de dos años. ###

Formato de salida: JSON lista de objetos."""
print(ask_llm(prompt_caso1))
```

### Caso 2: Soporte Técnico Automatizado
**Técnicas usadas:** Role Prompting + Give an 'out'.
**Justificación:** Definir el rol asegura un tono profesional y el 'out' evita que el soporte invente soluciones técnicas peligrosas.

```python
prompt_caso2 = """Eres un asistente de soporte técnico de routers. 
Usa el manual: 'Si la luz es roja, reinicia. Si es azul, todo está bien'.
Si te preguntan por luces de otros colores, di que deben llamar al técnico.

Pregunta del usuario: 'Mi router tiene una luz verde parpadeando, ¿qué hago?'"""
print(ask_llm(prompt_caso2))
```

## Conclusiones

Tras probar estas técnicas, me he dado cuenta de que el diseño del prompt es lo que realmente marca la diferencia en la calidad de los resultados.

Las técnicas que más me han servido son Few-shot, porque es mucho más rápido dar un par de ejemplos que escribir una instrucción larga, y Chain-of-Thought, porque ayuda a que el modelo no se equivoque en tareas lógicas al obligarlo a pensar paso a paso.

También destacaría lo útil que es forzar el formato (JSON) y dar una vía de escape (out) al modelo. La primera permite que la respuesta sea útil para ser usada por otros programas y la segunda evita que la IA invente datos cuando no sabe la respuesta, lo que hace que el sistema sea mucho más fiable. En definitiva, un prompt bien estructurado ahorra mucho trabajo de corrección posterior.
