# 🎓 Tutorial Avanzado: Técnicas de Prompt Engineering

## 🛠️ 1. Configuración del Laboratorio
Antes de empezar, cargamos nuestro entorno y definimos la función de conexión con nuestro endpoint de proyecto en **Azure AI Foundry**.

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

client = OpenAI(
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),
    base_url=os.getenv("AZURE_OPENAI_ENDPOINT")
)

def ask_tutorial(prompt):
    deployment = os.getenv("AZURE_OPENAI_DEPLOYMENT_NAME")
    res = client.chat.completions.create(
        model=deployment,
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7
    )
    return res.choices[0].message.content
```

### Self-Consistency (Auto-consistencia)
**¿Qué es?** Es una evolución del *Chain-of-Thought*. En lugar de generar una sola respuesta, se le pide al modelo que genere varios caminos de razonamiento y luego se selecciona la conclusión más repetida (voto mayoritario).

**¿Cuándo usarla?** En problemas matemáticos complejos o de lógica donde el modelo suele fallar en el primer intento.

```python
prompt_self_consistency = """Resuelve este problema de lógica de 3 formas distintas y separadas.
Problema: Si 5 máquinas tardan 5 minutos en hacer 5 widgets, ¿cuánto tardarían 100 máquinas en hacer 100 widgets?

Escribe 3 razonamientos numerados y al final indica la 'Respuesta Final'."""

print(ask_tutorial(prompt_self_consistency))
```

### Emotional Stimuli (Estímulos Emocionales)
**¿Qué es?** Investigaciones recientes demuestran que añadir frases que impliquen urgencia o importancia social (como "Esto es vital para mi carrera") mejora el rendimiento de los modelos en tareas de razonamiento.

**¿Cuándo usarla?** Cuando el modelo da respuestas perezosas o demasiado breves.

```python
prompt_emotion = """Explica cómo funciona la computación cuántica de forma que un principiante lo entienda.
Ten en cuenta que este es mi examen final y mi carrera depende de que tu explicación sea perfecta y detallada. 
¡Hazlo lo mejor posible!"""

print(ask_tutorial(prompt_emotion))
```

## Conclusiones del Tutorial
1. **La IA es sensible al contexto:** No solo importa lo que pides, sino cómo lo enmarcas.
2. **Iteración:** Si una técnica no funciona, prueba a combinarla con otra (ej. Persona + Self-Consistency).
3. **Prueba y Error:** El Prompt Engineering es una disciplina experimental; no hay un 'prompt perfecto' universal.
